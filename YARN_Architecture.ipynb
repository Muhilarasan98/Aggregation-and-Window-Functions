====YARN Architecture

Spark on YARN

Hadoop -

HDFS - Namenode, Datanode
MAPREDUCE - Processing computation
YARN - Resource manager

YARN --
Resource Manager (Master)
Node Manager (slave/worker)


you are invoking a hadoop job from client machine

hadoop jar <>

what will happen now?

The requests goes to the resource manager

Resource manager will coordinate with one of the Node Managers and create a container in that worker node

inside this container it will start a service called as Application Master 

this application master will act as a local manager to manage this application...



3 Containers

2 Containers 2 gb ram, 1 core, worker node 1
1 Container 1 gb ram, worker node 2

we got container on worker node 1 and node 2

node manager will come in

node manager manages the containers/executors which are running on worker nodes..

this application master will interact with the name node to understand where the blocks of file is kept in hdfs


uber mode - the job is soo small it can run in the container in which application master is running ,
it does not need other containers.

client mode - Notebook, spark shell - Interactive

cluster mode - spark submit 

client mode - the spark driver runs outside the spark cluster. it runs on the gateway node/edge node

cluter mode - the spark driver runs in the spark cluster on one of the worker machine, even if the gateway node crashes,

or even when we logout from gateway node the application will still run.. 

Driver is a heart of spark application -- when an driver crashes everything will get crashed.
